## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
air_bnb
````

````python
# code for getting the info and data type of variables
air_bnb.info()
````

````python
# code for dropping or removing the rows which has null values in variables name and host_name
air_bnb.dropna(subset=['name','host_name'],inplace=True)
````

````python
# code for dropping the column last review
air_bnb.drop(columns='last_review',inplace=True)
````

````python
# code for filling the null values of reviews per month of the listing
air_bnb.loc[:,'reviews_per_month'].fillna(value=0,inplace=True)
````

````python
#  code of boxplot for price variable to see outliers
sns.boxplot(air_bnb.loc[:,'price'])
````

````python
# code to check min pricing
air_bnb.loc[:,'price'].min()
````

````python
# code to remove listings which has zero pricing
air_bnb=air_bnb[air_bnb.loc[:,'price']!=0]
````

````python
# code to check the shape of dataset after imputation and dropping the null values and outliers
air_bnb.shape
````

````python
# code to check min pricing after dropping zero price listings
air_bnb.loc[:,'price'].min()
````

````python
# box plot code to see the price distribution
sns.boxplot(air_bnb.loc[:,'price'])
plt.grid(axis='y')
plt.ylim(0,200)
````

### What all manipulations have you done and insights you found?

1. During the data wrangling, it was found that missing values are there in 4 columns listing name, host name, last review and  reviews per month of the lsiting.

2. Removed the rows having null or missing values in listing name and host name columns as they cannot be imputed and has negligible number of missing values.

3. Removed last review column as it will not be used during analysis and has large number of missing values (approx. 20% of values are missing)

4. Filled missing values in reviews per month column with zero as we can assume that no review has been given or the  listing is new and hasn’t been reviewed yet.

5. From the box plot and mathematical operating we can observe price column has outliers and zero pricing values which doesn’t make any sense. So zero priced listings are deleted from dataset.

6. After imputation and dropping we have 48,847 rows and 15 columns in the data sets without any missing or null values.

7. From the box plot song we can observe that the median value(most of prices) is 105 and 50% of  prices are in the range of 60-175.
